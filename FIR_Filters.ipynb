# Low Pass Filter

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos,pi

# STEP 1: Set the cutoff Frequencies
fc = 5000  #upper analog cutoff frequency
fs = 16000  #sampling frequency

#STEP 2: Find the digital cutoff frequencies
wc = 2*pi*fc/fs   #upper digital cutoff frequency

#STEP 3: Select appropriate window,usually hamming window with suitable number of samples(N)
N = 65
hamming = 0.54 + 0.46 * cos(2*pi*np.arange(-(N-1)/2,(N-1)/2 +1)/(N-1))
plt.stem(np.arange(-(N-1)/2,(N-1)/2 +1),hamming);
plt.title('hamming window');
plt.show();

# STEP 4: Generate a signal with some random frequencies
f1 = 1500
f2 = 3000
f3 = 6000
f4 = 4500
n = np.arange(0,10/f1,1/fs)
x = sin(2*pi*f1*n) + sin(2*pi*f2*n) + sin(2*pi*f3*n) + sin(2*pi*f4*n)
plt.plot(np.arange(len(n)),x);
plt.title('Signal');
plt.xlabel('n');
plt.show();

#STEP 5 : Find the h[n] of the system/filter and multiply it by the window
hd = np.zeros((N))
alpha = (N-1)/2

for n in range(N):
    if n == alpha:
        hd[n] = (wc/pi)
    else:
        hd[n] = sin(wc*(n-alpha))/(pi*(n-alpha))
                
h = hd * hamming
plt.stem(np.arange(-(N-1)/2,(N-1)/2 +1),h);
plt.title('Filter h[n]');
plt.show();

#STEP 6: Find the convolution ,y[n] of  x[n] with h[n] (ie, signal generated and the filter coefficents)
y = np.convolve(x,h)
plt.stem(y);
plt.title('Filter output when x[n] is input.....y[n] = conv(x[n],h[h])');
plt.show();

#STEP 7: Find DFT of the signal to verify the frequencies
freq_axis = np.fft.fftfreq(len(x),1/fs)
plt.stem(freq_axis,np.fft.fft(x));
plt.title('DFT of x[n]');
plt.show();

#STEP 8: Compare the the DFT of X[n] with DFT of y[n]
freq_axis = np.fft.fftfreq(len(y),1/fs)
plt.stem(freq_axis,np.abs(np.fft.fft(y)));
plt.title('DFT of y[n]');
plt.show();
```

# Band Pass Filter

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos,pi

# STEP 1: Set the cutoff Frequencies
fc1 = 2000  #lower analog cutoff frequency
fc2 = 5000  #upper analog cutoff frequency
fs = 16000  #sampling frequency

#STEP 2: Find the digital cutoff frequencies
wc1 = 2*pi*fc1/fs   #lower digital cutoff frequency
wc2 = 2*pi*fc2/fs   #upper digital cutoff frequency

#STEP 3: Select appropriate window,usually hamming window with suitable number of samples(N)
N = 65
hamming = 0.54 + 0.46 * cos(2*pi*np.arange(-(N-1)/2,(N-1)/2 +1)/(N-1))
plt.stem(np.arange(-(N-1)/2,(N-1)/2 +1),hamming);
plt.title('hamming window');
plt.show();

# STEP 4: Generate a signal with some random frequencies
f1 = 3000
f2 = 1500
f3 = 6000
f4 = 4500
n = np.arange(0,10/f2,1/fs)
x = sin(2*pi*f1*n) + sin(2*pi*f2*n) + sin(2*pi*f3*n) + sin(2*pi*f4*n)
plt.plot(np.arange(len(n)),x);
plt.title('Signal');
plt.xlabel('n');
plt.show();

#STEP 5 : Find the h[n] of the system/filter and multiply it by the window
hd = np.zeros((N))
alpha = (N-1)/2

for n in range(N):
    if n == alpha:
        hd[n] = ((wc2-wc1)/pi)
    else:
        hd[n] = sin(wc2*(n-alpha))/(pi*(n-alpha))-sin(wc1*(n-alpha))/(pi*(n-alpha))
                
h = hd * hamming
plt.stem(np.arange(-(N-1)/2,(N-1)/2 +1),h);
plt.title('Filter h[n]');
plt.show();

#STEP 6: Find the convolution ,y[n] of  x[n] with h[n] (ie, signal generated and the filter coefficents)
y = np.convolve(x,h)
plt.stem(y);
plt.title('Filter output when x[n] is input.....y[n] = conv(x[n],h[h])');
plt.show();

#STEP 7: Find DFT of the signal to verify the frequencies
freq_axis = np.fft.fftfreq(len(x),1/fs)
plt.stem(freq_axis,np.fft.fft(x));
plt.title('DFT of x[n]');
plt.show();

#STEP 8: Compare the the DFT of X[n] with DFT of y[n]
freq_axis = np.fft.fftfreq(len(y),1/fs)
plt.stem(freq_axis,np.abs(np.fft.fft(y)));
plt.title('DFT of y[n]');
plt.show();
```
